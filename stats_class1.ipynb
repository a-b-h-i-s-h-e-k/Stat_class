# Mean Mode Median

### Mean: The Arithmetic Average
The mean, or average, is calculated by summing all the values in a dataset and dividing by the total
number of values.

### Median: The Middle Value
The median is the middle value in a sorted dataset. It divides the data into two halves, with half the
values being higher and the other half being lower.

ex : 13, 21, 21, *40*, 48, 55, 72
 
### Mode: The Most Frequent Value
The mode is the value that appears most frequently in a dataset. It represents the most common or
typical value in the data.
A dataset can have multiple modes if multiple values occur with the same highest frequenc

```python
import numpy as np
from scipy import stats

# Example array
data = np.array([1, 2, 2, 3, 4, 4, 4, 5, 6])

# Mean
mean_value = np.mean(data)

# Median
median_value = np.median(data)

# Mode
mode_value = stats.mode(data)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

# Range


The range is the difference between the smallest and the largest value of the data.

Range is the simplest measure of variation.

Here is a histogram of the age of all 934 Nobel Prize winners up to the year 2020, showing the range:

```python
# ptp() method to find the range
import numpy as np

values = [13,21,21,40,48,55,72]

x = np.ptp(values)

print(x)
```

```python
min_a = min(values)
max_a =max(values)
max_a-min_a
```

# Percentiles
## Percentiles are values that separate the data into 100 equal parts.

For example, The 95th percentile separates the lowest 95% of the values from the top 5%

The 25th percentile (P25%) is the same as the first quartile (Q1).

The 50th percentile (P50%) is the same as the second quartile (Q2) and the median.

The 75th percentile (P75%) is the same as the third quartile (Q3)

```python
import numpy

values = [13,21,21,40,42,48,55,72]

x = numpy.percentile(values, 25)

print(x)
```

# Standard Deviation


Standard deviation is the most commonly used measure of variation, which describes how spread out the data is.

Standard Deviation
Standard deviation () measures how far a 'typical' observation is from the average of the data ().

Standard deviation is important for many statistical methods.

Here is a histogram of the age of all 934 Nobel Prize winners up to the year 2020, showing standard deviations:

```python
from IPython.core.display import Image, display
display(Image(url="https://www.w3schools.com/statistics/img_histogram_std.svg", width=400, unconfined=True))
```

* Roughly 68.3% of the data is within 1 standard deviation of the average (from -1 to +1)
* Roughly 95.5% of the data is within 2 standard deviations of the average (from -2 to +2)
* Roughly 99.7% of the data is within 3 standard deviations of the average (from -3 to +3)


Note: A normal distribution has a "bell" shape and spreads out equally on both sides.

```python
import numpy

values = [4,11,7,14]

x = numpy.std(values)

print(x)
```
