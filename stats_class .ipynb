#  Center Limit Theorom

The Central Limit Theorem (CLT) is like a magic rule in statistics that helps us understand how data behaves when we take samples from a larger group (population). Here’s the idea in simple terms:

#### What it Says: 
If you take lots of random samples from a population and calculate the average (mean) of each sample, those averages will start to form a shape that looks like a bell curve, which is known as a normal distribution. This happens even if the original data doesn’t follow a bell curve.

#### Why It’s Useful:
The CLT is really handy because it lets us make guesses about the whole population based on just a sample. Since the sample means form a bell curve, we can use that curve to make predictions, test ideas, and estimate things about the population with confidence.

#### Why It Matters:
Understanding the CLT helps us make better decisions based on data. It’s a key reason why we can trust statistics and use them to draw conclusions, even when we’re only working with a part of the whole group.

In short, the Central Limit Theorem is a powerful tool that helps us make sense of data and draw conclusions, even if we don’t know much about the entire population.

# Lognormal Dist

* Not Bell Shape
* Used to Scale Data In Diffrent Form

# Ex

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data from a log-normal distribution
mu, sigma = 0, 1  # parameters for the underlying normal distribution
log_normal_data = np.random.lognormal(mu, sigma, 1000)

# Plot the log-normal distribution
plt.hist(log_normal_data, bins=30, density=True, alpha=0.6, color='g')
plt.title("Log-Normal Distribution")
plt.show()
```

# Log Normal TO Normal Distibution

```python
# Apply the natural logarithm to convert to a normal distribution
normal_data = np.log(log_normal_data)

# Plot the resulting normal distribution
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='b')
plt.title("Converted Normal Distribution")
plt.show()
```

# Hypithesis Testing

Suppose we have a sample of data representing the test scores of students in a class. We want to test whether the average score of this class is significantly different from a known population mean score of 75.

#### 1 ) Set up Hypotheses:
* Null Hypothesis (H0): The mean score of the class is equal to 75.
* Alternative Hypothesis (H1): The mean score of the class is not equal to 75.

#### 2) Calculate the Test Statistic:
* We’ll use the t-test to compare the sample mean to the population mean.

#### 3)Determine the p-value:
* The p-value will tell us whether to reject the null hypothesis.

#### 4) Make a Decision:
* If the p-value is less than the significance level (commonly 0.05), reject the null hypothesis.

```python
import numpy as np
import scipy.stats as stats

# Generate a sample of test scores
np.random.seed(0)  # for reproducibility
sample_data = np.random.normal(loc=73, scale=10, size=30)

# Known population mean
population_mean = 75
```

```python
# Perform the t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# Output the t-statistic and p-value
t_statistic, p_value
```

```python
# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print(f"Reject the null hypothesis. p-value = {p_value}")
else:
    print(f"Fail to reject the null hypothesis. p-value = {p_value}")
```

#### Explanation of Output

* T-statistic: This value tells us how far the sample mean is from the population mean in terms of standard errors.
* P-value: This is the probability of obtaining a test statistic as extreme as the one observed, under the assumption that the null hypothesis is true.

#### Interpretation

* If the p-value is less than 0.05, we reject the null hypothesis, concluding that the mean test score of the class is significantly different from 75.
 * If the p-value is greater than 0.05, we fail to reject the null hypothesis, suggesting there isn’t enough evidence to say the class mean is different from 75.

# Annova Test

Suppose we want to test whether there is a significant difference in the average scores of students from three different study groups.
Steps:

* Create data: We have three study groups, and we'll create some sample data representing their scores.
* Perform ANOVA: Use the f_oneway function to perform the ANOVA test.

```python
# Sample data: Scores from three different study groups
group1 = [88, 92, 85, 91, 89]
group2 = [75, 78, 74, 80, 79]
group3 = [93, 95, 90, 97, 94]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Display the results
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# Conclusion
if p_value < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```

# Chi Square Test
A Chi-Square test is used to determine if there is a significant association between categorical variables. 

### Example: Chi-Square Test of Independence

Suppose we want to test whether there is an association between gender (male/female) and preference for a product (like/dislike).
### Steps:

*    Create a contingency table: This table shows the frequency of each combination of categories.
*    Perform the Chi-Square test: Use the chi2_contingency function from scipy.stats.

```python
import scipy.stats as stats
import numpy as np

# Contingency table: rows represent gender, columns represent preference
# Example data: [Male, Female] vs [Like, Dislike]
data = np.array([[30, 10],  # Males: 30 like, 10 dislike
                 [20, 40]]) # Females: 20 like, 40 dislike

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

# Display the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Conclusion
if p_value < 0.05:
    print("There is a significant association between gender and preference.")
else:
    print("There is no significant association between gender and preference.")
```
